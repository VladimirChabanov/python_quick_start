### Возможное решение

Т.к. данные линейно разделимы относительно интересующего нас показателя (`repaid`), то для решения этой задачи можно воспользоваться одним искусственным нейроном.

#### 1. Общая информация



Искусственный нейрон состоит из: $m$ входов, одного выхода и тела. Входные сигналы $x_i$ преобразуются в выходной $y$ в 2 шага:

- вычисляется $z$, как сумма всех входных сигналов $x_i$ умноженных на соответствующий весовой коэффициент $w_i$ и минус порог $b$ (bias):

$$ z = \sum\limits_{i=1}^m x_i \cdot w_i - b$$

- получившееся значение $z$ передаётся функции активации $\phi$:

$$ y = \phi(z) $$

**Примечание**

Иногда, в литературе, в формуле для расчёта $z$ можно встретить плюс перед $b$ или $b$ может вообще отсутствовать, но суть от этого никак не меняется. Разница только в обозначениях.

Т.е. мы можем представить $b$, как ещё один вход, с постоянным сигналом $x_0 = -1$ и весовым коэффициентом $w_0$: $b = x_0 \cdot w_0$. В таком представлении можно загнать $b$ под знак суммы, но нумерацию при этом начать с 0, а не с 1.

Смена, в формуле, знака перед $b$ с минуса на плюс, тоже не на что не повлияет, т.к. $b$ - это значение, которое будет нами вычисляться, а значит $b$ просто рассчитается с противоположным знаком, тем самым компенсируя смену знака в формуле.

Почему в данном случае выбрана форма записи с выделенным $b$ и знаком минус, станет понятно чуть дальше.

#### 2. Функция активации

В зависимости от задачи используются различные функции активации. В нашем случае, лучше всего подходит - пороговая. Пороговая функция активации устанавливает на выходе нейрона значение 1, в том случае, если $z \ge 0$ иначе на выходе нейрона будет 0 (первый вариант на рисунке).



Если подставить в уравнение $z \ge 0$ значение $z$ в явном виде, то получим:

$$ \sum\limits_{i=1}^m x_i \cdot w_i - b \ge 0$$

$$ \sum\limits_{i=1}^m x_i \cdot w_i \ge b$$

Т.е если взвешенная сумма входных сигналов превышает значение $b$, то на выходе нейрона 1 иначе 0. Именно поэтому $b$ называют порогом, отделяют от остальных весов и выбирают знак минус в уравнении. 


#### 3. Обучение

Обучение нейрона заключается в подборе таких значений весовых коэффициентов $w_i$ и порога $b$, чтобы набор входных сигналов $x_i$, был правильно преобразован в выходной сигнал $y$, для всего датасета.

Методов обучения существует множество. В нашем случае подойдёт простейший метод, который называется Дельта-правило.

Суть метода заключается в следующем. Для каждого набора входных сигналов $x$ нам известно правильное значение выходного сигнала $d$. Если нейрон, для текущих входов выдаёт правильный выход $d - y = 0$, то мы ничего не делаем. Если выход получился не правильный, то корректируем весовые коэффициенты (порог тоже), так, чтобы подвинуть выход в правильную сторону. Направление, в котором нужно подвинуть вес $w_i$ определяется входным сигналом $x_i$ и знаком ошибки выхода.

Разница между правильным выходом $d$ и тем, что у нас получилось $y$ называется дельта:

$$\Delta = d - y$$

Для каждого набора вход-выход весовые коэффициенты корректируются по формуле:

$$ w_i(new) = w_i(old) + x_i \cdot \Delta $$

$$ b(new) = b(old) + (-1) \cdot \Delta $$

Процесс обучения продолжается до тех пор, пока $\Delta$ не равна нулю, хотя бы для одной пары вход-выход.

Обычно в эти формулы добавляют специальный коэффициент $k$, который позволяет управлять скоростью обучения. Его так и называют $k$ - коэффициент скорости обучения. Процесс обучения зависит от того, как сильно мы изменяем весовые коэффициенты $w_i$ и $b$, поэтому формула с добавленным коэффициентом обучения будет выглядеть так:

$$ w_i(new) = w_i(old) + x_i \cdot \Delta \cdot k $$

$$ b(new) = b(old) + (-1) \cdot \Delta  \cdot k $$

Коэффициент $k$ подбирают вручную под разные задачи. $k \ge 1$ - ускоряет обучение, а $0 < k \le 1$ - замедляет.

Большие коэффициенты $k$ могут привести к тому, что обучение никогда не остановится, т.к. изменения весов будут слишком сильные и правильные значения постоянно будут проскакиваться.

Маленькие коэффициенты $k$ хоть и могут замедлить обучение, но делают его стабильнее.

#### 4. Использование

Нейрон считается обученным, когда нам известны значения весов (и порога) которые правильно преобразовывают входы в выходы.

Теперь, для заданных входных значений $x_i$ мы можем получить значение на выходе $y$ по формулам:

$$ z = \sum\limits_{i=1}^m x_i \cdot w_i - b$$

$$ y = 1 \: if \: z \ge 0 \: else \: 0 $$


#### 5. Геометрический смысл и визуализация

Если посмотреть на уравнение:

$$ \sum\limits_{i=1}^m x_i \cdot w_i - b = 0$$

с точки зрения геометрии, то можно понять, что это уравнение гиперплоскости в $m$ мерном пространстве.

Для простоты, ограничимся случаем нейрона с 2мя входами. Тогда формула примет вид:

$$ x_1 \cdot w_1 + x_2 \cdot w_2 + (-1) \cdot b = 0$$

это [уравнение прямой](https://ru.wikipedia.org/wiki/Прямая#Уравнения_прямой_на_плоскости) в 2D, а $x_1$ и $x_2$ - это оси координат (как бы $x$ и $y$). Соответственно для 3х входов была бы плоскость в 3D и так далее.

В процессе обучения происходит изменение весовых коэффициентов, что в свою очередь приводит к вращению и смещению прямой построенной по этому уравнению таким образом, чтобы одна часть данных оказалась по одну сторону от прямой, а вторая по другую.

Отсюда и требование на линейную разделимость данных. Т.е. один нейрон в принципе не способен решить задачу, в которой данные нельзя разделить на две группы при помощи прямой линии (плоскости и т.д.).

#### 6. Пример

Пусть у нас есть данные о трёх клиентах:

salary | request | repaid
:-:    | :-:     | :-:
100    | 10      | 1
100    | 100     | 0
500    | 100     | 1

В этом случае нам нужен нейрон с двумя входами, по одному для  `salary` и `request` и одним выходом (он всегда один) для `repaid`.

Выберем начальное значение весовых коэффициентов $ws$ = 0, $wr$ = 1 и $b$ = 1 (обычно эти значения выбираются случайным образом) и коэффициента скорости обучения $k$ = 0.01.

1. Вычислим $z$, а затем $y$ для первого клиента:

  $$z = 100 \cdot ws + 10 \cdot wr - b = 0 + 10 - 1 = 9$$

  $$y = 1 \: if \: 9 \ge 0 \: else \: 0 = 1$$

  $$\Delta = d - y = 1 - 1 = 0$$

  Значит, для первого клиента корректировка весов НЕ нужна.

2. Вычислим $z$, а затем $y$ для второго клиента:

  $$z = 100 \cdot ws + 100 \cdot wr - b = 0 + 100 - 1 = 99$$

  $$y = 1 \: if \: 99 \ge 0 \: else \: 0 = 1$$

  $$\Delta = d - y = 0 - 1 = -1$$

  Значит, для второго клиента корректировка весов нужна.

  Корректируем веса:

  $$ws = ws + salary_2 \cdot \Delta \cdot k = 0 + 100 \cdot (-1) \cdot 0.01 = -1$$

  $$wr = wr + request_2 \cdot \Delta \cdot k = 1 + 100 \cdot (-1) \cdot 0.01 = 0$$

  $$b = b + (-1) \cdot \Delta \cdot k = 1 + (-1) \cdot (-1) \cdot 0.01 = 1.01$$

3. Вычислим $z$, а затем $y$ для третьего клиента (с новыми значениями весов):

  $$z = 500 \cdot ws + 100 \cdot wr - b = 500 \cdot (-1) + 0 - 1.01 = -501.01$$

  $$y = 1 \: if \: (-501.01) \ge 0 \: else \: 0 = 0$$

  $$\Delta = d - y = 1 - 0 = 1$$

  Значит, для третьего клиента корректировка весов нужна.

  Корректируем веса:

  $$ws = ws + salary_3 \cdot \Delta \cdot k = -1 + 500 \cdot 1 \cdot 0.01 = 4$$

  $$wr = wr + request_3 \cdot \Delta \cdot k = 0 + 100 \cdot 1 \cdot 0.01 = 1$$

  $$b = b + (-1) \cdot \Delta \cdot k = 1.01 + (-1) \cdot 1 \cdot 0.01 = 1$$

Первый проход по данным завершён. Мы видим, что $\Delta$ для некоторых клиентов не равно 0, а значит, процесс обучения не завершён. Идём на второй круг.

В результате нейрон обучится за 15 итераций. Веса будут меняться следующим образом:

ws | wr | b
:-:|:-:|:-:
0.0 | 1.0 | 1.0
-1.0 | 0.0 | 1.01
4.0 | 1.0 | 1.0
4.0 | 1.0 | 1.0
3.0 | 0.0 | 1.01
3.0 | 0.0 | 1.01
